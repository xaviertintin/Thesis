# REANA Classic
Transform CSV Format

```python
import csv
from datetime import datetime

# Define the input and output file paths
input_file_path = 'R-K0-1_cpu.csv'
output_file_path = 'R-K0-1_cpu_output.csv'

# Function to convert timestamp to the desired format
def convert_timestamp(timestamp):
    # Remove the time zone information
    timestamp = timestamp.split(' GMT')[0]
    # Parse the date without time zone information
    dt = datetime.strptime(timestamp, '%a %b %d %Y %H:%M:%S')
    return dt.strftime('%H:%M:%S')

# Open the input CSV file for reading and the output CSV file for writing
with open(input_file_path, mode='r') as infile, open(output_file_path, mode='w', newline='') as outfile:
    reader = csv.reader(infile)
    writer = csv.writer(outfile)

    # Process each row in the input file
    for row in reader:
        # Convert the timestamp and write the modified row to the output file
        converted_row = [convert_timestamp(row[0]), row[1]]
        writer.writerow(converted_row)

print(f'Converted data has been saved to {output_file_path}')
```

## CPU Usage

```python
import csv
import matplotlib.pyplot as plt
from datetime import datetime

# Define the input file path
input_file_path = 'R-K0-1_cpu_output.csv'

# Initialize lists to store the data
times = []
values = []

# Read the CSV file and process the data
with open(input_file_path, mode='r') as infile:
    reader = csv.reader(infile)
    # Read the first row to determine the start time
    first_row = next(reader)
    start_time = datetime.strptime(first_row[0], '%H:%M:%S')
    start_value = float(first_row[1])
    
    # Append the start time and value
    times.append(0)  # Start time is 0 minutes
    values.append(start_value)
    
    # Process the remaining rows
    for row in reader:
        # Parse the time and value
        time_str = row[0]
        value = float(row[1])
        
        # Convert time to minutes since start
        time_dt = datetime.strptime(time_str, '%H:%M:%S')
        minutes_since_start = (time_dt - start_time).seconds / 60
        
        # Append to the lists
        times.append(minutes_since_start)
        values.append(value)

# Plot the data
plt.figure(figsize=(10, 5))
plt.plot(times, values, marker='o')
plt.xlabel('Time (minutes)')
plt.ylabel('Value')
plt.title('CPU Usage')
plt.grid(True)
plt.show()
```

# CPU Usage Comparison

```python
import csv
import matplotlib.pyplot as plt
from datetime import datetime

# Function to read and process CSV data
def read_csv_data(file_path):
    times = []
    values = []
    with open(file_path, mode='r') as infile:
        reader = csv.reader(infile)
        # Read the first row to determine the start time
        first_row = next(reader)
        start_time = datetime.strptime(first_row[0], '%H:%M:%S')
        start_value = float(first_row[1])
        
        # Append the start time and value
        times.append(0)  # Start time is 0 minutes
        values.append(start_value)
        
        # Process the remaining rows
        for row in reader:
            # Parse the time and value
            time_str = row[0]
            value = float(row[1])
            
            # Convert time to minutes since start
            time_dt = datetime.strptime(time_str, '%H:%M:%S')
            minutes_since_start = (time_dt - start_time).seconds / 60
            
            # Append to the lists
            times.append(minutes_since_start)
            values.append(value)
    
    return times, values

# Define the input file paths
input_file_path1 = 'R-K0-1_cpu_output.csv'
input_file_path2 = 'R-K1-1_cpu_output.csv' 

# Read and process both CSV files
times1, values1 = read_csv_data(input_file_path1)
times2, values2 = read_csv_data(input_file_path2)

# Plot the data
plt.figure(figsize=(10, 5))
plt.plot(times1, values1, marker='o', label='REANA Classic')
plt.plot(times2, values2, marker='x', label='REANA Kueue')
plt.xlabel('Time (minutes)')
plt.ylabel('CPU Percentage')
plt.title('CPU Usage Comparison - 1 Concurrent Workflow')
plt.legend()
plt.grid(True)
plt.show()
```
