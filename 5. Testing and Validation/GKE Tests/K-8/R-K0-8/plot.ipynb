```python
import pandas as pd
import matplotlib.pyplot as plt

# Read CSV files
df_batch = pd.read_csv("reana_batch_job_counts.csv")
df_job = pd.read_csv("reana_job_counts.csv")

# Convert timestamp to datetime
df_batch['Timestamp'] = pd.to_datetime(df_batch['Timestamp'])
df_job['Timestamp'] = pd.to_datetime(df_job['Timestamp'])

# Plotting
plt.figure(figsize=(12, 6))

# Plot batch jobs
plt.plot(df_batch['Timestamp'], df_batch['Created'], label='Batch Jobs (Created)', linestyle='-')
plt.plot(df_batch['Timestamp'], df_batch['Pending'], label='Batch Jobs (Pending)', linestyle='--')
plt.plot(df_batch['Timestamp'], df_batch['Running'], label='Batch Jobs (Running)', linestyle='-.')

# Plot regular jobs
plt.plot(df_job['Timestamp'], df_job['Created'], label='Regular Jobs (Created)', linestyle='-')
plt.plot(df_job['Timestamp'], df_job['Pending'], label='Regular Jobs (Pending)', linestyle='--')
plt.plot(df_job['Timestamp'], df_job['Running'], label='Regular Jobs (Running)', linestyle='-.')

# Add labels and legend
plt.xlabel('Time')
plt.ylabel('Count')
plt.title('Job Status Over Time')
plt.legend()

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Display plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read CSV files
df_batch = pd.read_csv("reana_batch_pod_counts.csv")
df_job = pd.read_csv("reana_job_pod_counts.csv")

# Convert timestamp to datetime
df_batch['Timestamp'] = pd.to_datetime(df_batch['Timestamp'])
df_job['Timestamp'] = pd.to_datetime(df_job['Timestamp'])

# Plotting
plt.figure(figsize=(12, 6))

# Plot batch pods
plt.plot(df_batch['Timestamp'], df_batch['Pending'], label='Batch Pods (Pending)', linestyle='--')
plt.plot(df_batch['Timestamp'], df_batch['Running'], label='Batch Pods (Running)', linestyle='-')
plt.plot(df_batch['Timestamp'], df_batch['Succeeded'], label='Batch Pods (Succeeded)', linestyle='-.')

# Plot regular pods
plt.plot(df_job['Timestamp'], df_job['Pending'], label='Regular Pods (Pending)', linestyle='--')
plt.plot(df_job['Timestamp'], df_job['Running'], label='Regular Pods (Running)', linestyle='-')
plt.plot(df_job['Timestamp'], df_job['Succeeded'], label='Regular Pods (Succeeded)', linestyle='-.')

# Add labels and legend
plt.xlabel('Time')
plt.ylabel('Count')
plt.title('Pod Status Over Time')
plt.legend()

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Display plot
plt.tight_layout()
plt.show()
```
